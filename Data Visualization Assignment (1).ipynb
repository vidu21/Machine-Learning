## Q.9 


- Plot the Pie Chart for the two colums with some explode for the country "India"  and proper shadow

```python
d1={'Country':['United States','Great Britain','India','Russia','Germany'],
   'Medal':[46,27,26,19,17]}
df=pd.DataFrame(d1)
df.head()
```

```python
country_data = df["Country"]
medal_data = df["Medal"]
colors = ["#1f77b4", "#ff7f0e", "#2ca02c", "#d62728", "#8c564b"]
explode = (0.1, 0, 0, 0, 0) 
plt.pie(medal_data, labels=country_data, explode=explode, colors=colors,
autopct='%1.1f%%', shadow=True, startangle=140)
plt.title("Medal")
plt.show()
```
